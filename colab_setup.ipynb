```python
# ngrokのトークンを設定（https://ngrok.com/ でアカウント作成してトークンを取得）
from pyngrok import ngrok

# ここにngrokのトークンを入力
ngrok_token = "YOUR_NGROK_TOKEN_HERE"
ngrok.set_auth_token(ngrok_token)
```

```python
# Streamlitアプリを起動
import subprocess
import threading

def run_streamlit():
    subprocess.run(["streamlit", "run", "main.py", "--server.port", "8501"])

# バックグラウンドで実行
thread = threading.Thread(target=run_streamlit, daemon=True)
thread.start()

print("Streamlitアプリを起動中...")
```

```python
# ngrokで公開URLを取得
import time
time.sleep(5)  # Streamlitの起動を待つ

public_url = ngrok.connect(8501)
print("\n" + "="*50)
print("アプリのURL:")
print(public_url)
print("="*50)
print("\nこのURLをブラウザで開いてください。")
print("注意: ngrokの無料プランでは、セッションは一時的です。")
```
